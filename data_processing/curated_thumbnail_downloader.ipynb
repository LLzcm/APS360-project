```python
def get_thumbnails(channel_name, channel_id,category,vid_limit, train=True):
  val_counter = 0
  train_counter = 0
  channel_info = api.get_channel_info(channel_id=channel_id)
  uploads_playlist = channel_info.items[0].to_dict()['contentDetails']['relatedPlaylists']['uploads']
  playlist_videos = api.get_playlist_items(playlist_id=uploads_playlist, count=None)

  for k, video in enumerate(playlist_videos.items):
    if train is False:
      path = test_path
    elif val_counter < 10:
      path = val_path
    else:
      path = train_path
    if train_counter == vid_limit:
      break
    try:
      videoId = video.snippet.resourceId.videoId
      thumbnail_url = video.to_dict()['snippet']['thumbnails']['maxres']['url']
      savepath = path+'/'+category
      if(os.path.exists(savepath) is False):
        os.mkdir(savepath)
      opened = urlopen(thumbnail_url)
      image_array = plt.imread(opened, 0)
      im=Image.fromarray(image_array)
      savepath = path+'/'+category+'/'+videoId+'.jpg'
      if(os.path.isfile(savepath) is True):
        continue
      im.save(savepath)
      if val_counter < 10:
        val_counter += 1
      else:
        train_counter += 1
    except:
      print(k)
```

```python
def get_test_thumbnails(channel_name, channel_id,category,vid_limit):
  test_counter = 0
  channel_info = api.get_channel_info(channel_id=channel_id)
  uploads_playlist = channel_info.items[0].to_dict()['contentDetails']['relatedPlaylists']['uploads']
  playlist_videos = api.get_playlist_items(playlist_id=uploads_playlist, count=None)

  for k, video in enumerate(playlist_videos.items):
    path = test_path
    try:
      if test_counter < (vid_limit/10):
        videoId = video.snippet.resourceId.videoId
        thumbnail_url = video.to_dict()['snippet']['thumbnails']['maxres']['url']
        savepath = path+'/'+category
        if(os.path.exists(savepath) is False):
          os.mkdir(savepath)
        opened = urlopen(thumbnail_url)
        image_array = plt.imread(opened, 0)
        im=Image.fromarray(image_array)
        savepath = path+'/'+category+'/'+videoId+'.jpg'
        if(os.path.isfile(savepath) is True):
          continue
        im.save(savepath)
        test_counter += 1
      else:
        break
    except:
      print(k)
```
